```python
def lighttimedifference(v,prop):


    lightyear=scipy.constants.c /1000 *365*24*60*60#km
    covered_dist=v*365.25*24*60*60 /lightyear

    delta=covered_dist*prop

    return delta #mas
```

```python
def secular_acceleration(vr_kms, parallax_mas,mu_mas_per_yr):
    
    # Scalar formula for secular acceleration

    A=9.7779222168e8 #mas km yr /s
    dot_mu_mas_per_yr2 = -2*vr_kms*parallax_mas*mu_mas_per_yr /A
    
    return dot_mu_mas_per_yr2
```
